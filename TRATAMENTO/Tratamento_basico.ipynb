# Limpeza dos csv's


```python
import pandas as pd
import numpy as np

# =========================================================
# 1) CAMINHOS DOS ARQUIVOS
# =========================================================
arquivo_byd = "2026-02-04 16-43-40.csv"
arquivo_peugeot = "2026-02-05 10-33-45.csv"

saida_byd = "byd_dolphin_tratado.csv"
saida_peugeot = "peugeot_2008_tratado.csv"

# =========================================================
# 2) CONFIGURAÇÕES
# =========================================================
THRESH_NAN_BYD = 0.98
THRESH_NAN_PEUGEOT = 0.95

# =========================================================
# 3) COLUNAS IMPORTANTES (PRESERVAR)
# =========================================================
important_cols_byd = [
    "time",
    "[BMS] Battery total current (A)",
    "[BMS] Battery total voltage (V)",
    "[VCU] Velocidade do veículo (km/h)",
    "Potência da bateria de alta tensão do veículo elétrico (hp)",
    "Aceleração (g)",
    "Velocidade (GPS) (km/h)",
    "Velocidade média (GPS) (km/h)",
    "Altitude (GPS) (m)",
    "Latitude",
    "Longtitude",
]

important_cols_peugeot = [
    "time",
    "[BMS] Battery DC Current (A)",
    "[BMS] Battery DC Voltage (V)",
    "[VCU] Velocidade do veículo (ESP) (km/h)",
    "Potência da bateria de alta tensão do veículo elétrico (hp)",
    "Aceleração (g)",
    "Velocidade (GPS) (km/h)",
    "Velocidade média (GPS) (km/h)",
    "Altitude (GPS) (m)",
    "Latitude",
    "Longitude",
    "[MCU] Accelerator pedal position (%)",
]

# =========================================================
# 4) FUNÇÃO DE LIMPEZA
# =========================================================
def tratar_csv(
    path_csv,
    thresh_nan=0.95,
    preserve_cols=None,
    rename_map=None,
    output_path=None,
    nome_dataset="dataset"
):
    preserve_cols = preserve_cols or []
    rename_map = rename_map or {}

    # Leitura
    df = pd.read_csv(path_csv)

    n_rows_ini, n_cols_ini = df.shape
    cols_iniciais = df.columns.tolist()

    # -----------------------------------------------------
    # A) Remover colunas Unnamed
    # -----------------------------------------------------
    unnamed_cols = [c for c in df.columns if str(c).startswith("Unnamed:")]
    unnamed_cols = [c for c in unnamed_cols if c not in preserve_cols]
    df = df.drop(columns=unnamed_cols, errors="ignore")

    # -----------------------------------------------------
    # B) Remover colunas 100% NaN (exceto preservadas)
    # -----------------------------------------------------
    all_nan_cols = df.columns[df.isna().all()].tolist()
    all_nan_cols = [c for c in all_nan_cols if c not in preserve_cols]
    df = df.drop(columns=all_nan_cols, errors="ignore")

    # -----------------------------------------------------
    # C) Remover colunas muito esparsas (exceto preservadas)
    # -----------------------------------------------------
    nan_ratio = df.isna().mean()
    sparse_cols = nan_ratio[nan_ratio > thresh_nan].index.tolist()
    sparse_cols = [c for c in sparse_cols if c not in preserve_cols]
    df = df.drop(columns=sparse_cols, errors="ignore")

    # -----------------------------------------------------
    # D) Padronizar nomes de colunas principais
    # -----------------------------------------------------
    # Só renomeia o que existir no dataframe
    rename_map_valid = {k: v for k, v in rename_map.items() if k in df.columns}
    df = df.rename(columns=rename_map_valid)

    # -----------------------------------------------------
    # E) Tratar coluna de tempo
    # -----------------------------------------------------
    if "time" in df.columns:
        # tenta converter
        df["time"] = pd.to_datetime(df["time"], errors="coerce")

        # remove linhas sem tempo válido
        df = df[df["time"].notna()].copy()

        # ordena
        df = df.sort_values("time")

        # remove duplicatas de timestamp (mantém a última)
        df = df.drop_duplicates(subset=["time"], keep="last")

        # reset index
        df = df.reset_index(drop=True)

    # -----------------------------------------------------
    # F) Converter colunas numéricas mais prováveis
    # -----------------------------------------------------
    # tenta converter tudo que não for time para numérico quando possível
    for col in df.columns:
        if col != "time":
            df[col] = pd.to_numeric(df[col], errors="ignore")

    # -----------------------------------------------------
    # G) Salvar
    # -----------------------------------------------------
    if output_path is not None:
        df.to_csv(output_path, index=False)

    # -----------------------------------------------------
    # H) Resumo
    # -----------------------------------------------------
    n_rows_fim, n_cols_fim = df.shape
    cols_finais = df.columns.tolist()
    removidas = [c for c in cols_iniciais if c not in cols_finais]

    print("=" * 70)
    print(f"RESUMO - {nome_dataset}")
    print("=" * 70)
    print(f"Linhas iniciais          : {n_rows_ini}")
    print(f"Linhas finais            : {n_rows_fim}")
    print(f"Colunas iniciais         : {n_cols_ini}")
    print(f"Colunas finais           : {n_cols_fim}")
    print(f"Colunas removidas        : {len(removidas)}")
    print(f" - Unnamed removidas     : {len(unnamed_cols)}")
    print(f" - 100% NaN removidas    : {len(all_nan_cols)}")
    print(f" - > {thresh_nan:.0%} NaN removidas : {len(sparse_cols)}")

    if "time" in df.columns:
        print(f"Tempo mínimo             : {df['time'].min()}")
        print(f"Tempo máximo             : {df['time'].max()}")

    print("\nColunas finais:")
    print(df.columns.tolist())

    if output_path is not None:
        print(f"\nArquivo salvo em: {output_path}")

    return df

# =========================================================
# 5) MAPEAMENTO DE NOMES PARA ESQUEMA COMUM
# =========================================================
rename_byd = {
    "[BMS] Battery total current (A)": "current_A",
    "[BMS] Battery total voltage (V)": "voltage_V",
    "[VCU] Velocidade do veículo (km/h)": "speed_kmh",
    "Potência da bateria de alta tensão do veículo elétrico (hp)": "power_hp",
    "Aceleração (g)": "acc_g",
    "Velocidade (GPS) (km/h)": "gps_speed_kmh",
    "Velocidade média (GPS) (km/h)": "gps_avg_speed_kmh",
    "Altitude (GPS) (m)": "gps_altitude_m",
    "Latitude": "latitude",
    "Longtitude": "longitude",   # corrigindo nome inconsistente
}

rename_peugeot = {
    "[BMS] Battery DC Current (A)": "current_A",
    "[BMS] Battery DC Voltage (V)": "voltage_V",
    "[VCU] Velocidade do veículo (ESP) (km/h)": "speed_kmh",
    "Potência da bateria de alta tensão do veículo elétrico (hp)": "power_hp",
    "Aceleração (g)": "acc_g",
    "Velocidade (GPS) (km/h)": "gps_speed_kmh",
    "Velocidade média (GPS) (km/h)": "gps_avg_speed_kmh",
    "Altitude (GPS) (m)": "gps_altitude_m",
    "Latitude": "latitude",
    "Longitude": "longitude",
    "[MCU] Accelerator pedal position (%)": "pedal_real_pct",
}

# =========================================================
# 6) PROCESSAR OS DOIS ARQUIVOS
# =========================================================
df_byd_tratado = tratar_csv(
    path_csv=arquivo_byd,
    thresh_nan=THRESH_NAN_BYD,
    preserve_cols=important_cols_byd,
    rename_map=rename_byd,
    output_path=saida_byd,
    nome_dataset="BYD Dolphin - 2026-02-04"
)

df_peugeot_tratado = tratar_csv(
    path_csv=arquivo_peugeot,
    thresh_nan=THRESH_NAN_PEUGEOT,
    preserve_cols=important_cols_peugeot,
    rename_map=rename_peugeot,
    output_path=saida_peugeot,
    nome_dataset="Peugeot e-2008 - 2026-02-05"
)

# =========================================================
# 7) PRÉVIA DOS DADOS TRATADOS
# =========================================================
print("\n" + "=" * 70)
print("AMOSTRA BYD TRATADO")
print("=" * 70)
print(df_byd_tratado.head())

print("\n" + "=" * 70)
print("AMOSTRA PEUGEOT TRATADO")
print("=" * 70)
print(df_peugeot_tratado.head())
```

# Interpolação

```python
import pandas as pd
import numpy as np

# =========================================================
# 1) CAMINHOS DOS ARQUIVOS TRATADOS
# =========================================================
arquivo_byd_tratado = "byd_dolphin_tratado.csv"
arquivo_peugeot_tratado = "peugeot_2008_tratado.csv"

saida_byd_reamostrado = "byd_dolphin_reamostrado_v2.csv"
saida_peugeot_reamostrado = "peugeot_2008_reamostrado_v2.csv"

# =========================================================
# 2) CONFIGURAÇÕES
# =========================================================
FREQ = "250ms"
MAX_GAP_SECONDS = 2.0

COLUNAS_PRIORITARIAS = [
    "current_A",
    "voltage_V",
    "speed_kmh",
    "power_hp",
    "acc_g",
    "gps_speed_kmh",
    "gps_avg_speed_kmh",
    "gps_altitude_m",
    "latitude",
    "longitude",
    "pedal_real_pct",
]

# =========================================================
# 3) FUNÇÃO CORRIGIDA
# =========================================================
def reamostrar_e_interpolar_v2(
    path_csv,
    output_path,
    nome_dataset="dataset",
    freq="250ms",
    max_gap_seconds=2.0,
    colunas_prioritarias=None,
):
    colunas_prioritarias = colunas_prioritarias or []

    # ---------------------------------------------
    # A) Ler e preparar
    # ---------------------------------------------
    df = pd.read_csv(path_csv)

    if "time" not in df.columns:
        raise ValueError(f"O arquivo {path_csv} não possui a coluna 'time'.")

    df["time"] = pd.to_datetime(df["time"], errors="coerce")
    df = df[df["time"].notna()].copy()
    df = df.sort_values("time").drop_duplicates(subset=["time"], keep="last")
    df = df.reset_index(drop=True)

    # ---------------------------------------------
    # B) Selecionar colunas úteis existentes
    # ---------------------------------------------
    cols_existentes = [c for c in colunas_prioritarias if c in df.columns]

    for col in cols_existentes:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # ---------------------------------------------
    # C) Arredondar timestamps para a grade
    # ---------------------------------------------
    df["time_bin"] = df["time"].dt.round(freq)

    # marca se aquele bin veio de amostra original
    bins_originais = set(df["time_bin"])

    # ---------------------------------------------
    # D) Agregar por bin temporal
    # - média para sinais numéricos
    # ---------------------------------------------
    agg_dict = {col: "mean" for col in cols_existentes}

    df_bin = (
        df.groupby("time_bin", as_index=True)
          .agg(agg_dict)
          .sort_index()
    )
    df_bin.index.name = "time"

    # ---------------------------------------------
    # E) Expandir grade regular completa
    # ---------------------------------------------
    start_time = df_bin.index.min().floor(freq)
    end_time = df_bin.index.max().ceil(freq)
    novo_indice = pd.date_range(start=start_time, end=end_time, freq=freq)

    df_rs = df_bin.reindex(novo_indice)
    df_rs.index.name = "time"

    # ---------------------------------------------
    # F) Interpolar só gaps curtos
    # ---------------------------------------------
    step_seconds = pd.to_timedelta(freq).total_seconds()
    limit_steps = max(1, int(max_gap_seconds / step_seconds))

    for col in cols_existentes:
        df_rs[col] = df_rs[col].interpolate(
            method="time",
            limit=limit_steps,
            limit_direction="both",
            limit_area="inside",
        )

    # ---------------------------------------------
    # G) Marcar se esse timestamp corresponde a bin com amostra real
    # ---------------------------------------------
    df_rs["is_original_sample"] = df_rs.index.isin(bins_originais).astype(int)

    # ---------------------------------------------
    # H) Máscara de validade menos rígida
    # Regra:
    #   - precisa de current_A e speed_kmh
    #   - e pelo menos um entre power_hp, voltage_V ou acc_g
    # ---------------------------------------------
    valid_mask = pd.Series(True, index=df_rs.index)

    if "current_A" in df_rs.columns:
        valid_mask &= df_rs["current_A"].notna()
    else:
        valid_mask &= False

    if "speed_kmh" in df_rs.columns:
        valid_mask &= df_rs["speed_kmh"].notna()
    else:
        valid_mask &= False

    aux_mask = pd.Series(False, index=df_rs.index)

    if "power_hp" in df_rs.columns:
        aux_mask |= df_rs["power_hp"].notna()

    if "voltage_V" in df_rs.columns:
        aux_mask |= df_rs["voltage_V"].notna()

    if "acc_g" in df_rs.columns:
        aux_mask |= df_rs["acc_g"].notna()

    valid_mask &= aux_mask
    df_rs["valid_for_pedal_proxy"] = valid_mask.astype(int)

    # ---------------------------------------------
    # I) Estatísticas
    # ---------------------------------------------
    total_linhas = len(df_rs)
    pct_valid = 100 * df_rs["valid_for_pedal_proxy"].mean()

    print("=" * 72)
    print(f"RESUMO - {nome_dataset}")
    print("=" * 72)
    print(f"Frequência nova               : {freq}")
    print(f"Passo da grade (s)            : {step_seconds:.3f}")
    print(f"Máx. gap interpolado (s)      : {max_gap_seconds}")
    print(f"Limite em passos              : {limit_steps}")
    print(f"Linhas na grade final         : {total_linhas}")
    print(f"Linhas válidas p/ pedal proxy : {df_rs['valid_for_pedal_proxy'].sum()} ({pct_valid:.2f}%)")

    print("\nPercentual não nulo por coluna:")
    for col in cols_existentes:
        pct = 100 * df_rs[col].notna().mean()
        print(f" - {col}: {pct:.2f}%")

    # ---------------------------------------------
    # J) Salvar
    # ---------------------------------------------
    df_out = df_rs.reset_index()
    df_out.to_csv(output_path, index=False)

    print(f"\nArquivo salvo em: {output_path}")

    return df_out

# =========================================================
# 4) RODAR NOS DOIS ARQUIVOS
# =========================================================
df_byd_rs = reamostrar_e_interpolar_v2(
    path_csv=arquivo_byd_tratado,
    output_path=saida_byd_reamostrado,
    nome_dataset="BYD Dolphin - reamostrado v2",
    freq=FREQ,
    max_gap_seconds=MAX_GAP_SECONDS,
    colunas_prioritarias=COLUNAS_PRIORITARIAS,
)

print("\n")

df_peugeot_rs = reamostrar_e_interpolar_v2(
    path_csv=arquivo_peugeot_tratado,
    output_path=saida_peugeot_reamostrado,
    nome_dataset="Peugeot e-2008 - reamostrado v2",
    freq=FREQ,
    max_gap_seconds=MAX_GAP_SECONDS,
    colunas_prioritarias=COLUNAS_PRIORITARIAS,
)

# =========================================================
# 5) PRÉVIA
# =========================================================
print("\n" + "=" * 72)
print("AMOSTRA - BYD REAMOSTRADO V2")
print("=" * 72)
print(df_byd_rs.head(10))

print("\n" + "=" * 72)
print("AMOSTRA - PEUGEOT REAMOSTRADO V2")
print("=" * 72)
print(df_peugeot_rs.head(10))
```

```python
import pandas as pd
import numpy as np

# =========================================================
# 1) CAMINHOS
# =========================================================
arquivo_volks = "2026-02-05 10-29-45 volks.csv"
arquivo_csvlog = "CSVLog_20260204_163517.csv"

saida_volks = "volks_combustao_tratado.csv"
saida_csvlog = "csvlog_combustao_tratado.csv"

# =========================================================
# 2) FUNÇÕES AUXILIARES
# =========================================================
def normalizar_nome_coluna(c):
    """Remove espaços extras nas bordas."""
    return str(c).strip()

def robust_minmax(series, q_low=0.05, q_high=0.95):
    """Normalização robusta opcional (não usada no salvamento, mas útil depois)."""
    s = pd.to_numeric(series, errors="coerce")
    low = s.quantile(q_low)
    high = s.quantile(q_high)
    if pd.isna(low) or pd.isna(high) or high == low:
        return pd.Series(np.nan, index=s.index)
    out = (s - low) / (high - low)
    return out.clip(0, 1)

# =========================================================
# 3) TRATAMENTO DO VOLKS (arquivo muito esparso)
# =========================================================
def tratar_volks(path_in, path_out, thresh_nan=0.98):
    df = pd.read_csv(path_in)

    # -----------------------------
    # A) limpar nomes de colunas
    # -----------------------------
    df.columns = [normalizar_nome_coluna(c) for c in df.columns]

    # -----------------------------
    # B) remover colunas "Unnamed"
    # -----------------------------
    unnamed_cols = [c for c in df.columns if c.startswith("Unnamed:")]
    df = df.drop(columns=unnamed_cols, errors="ignore")

    # -----------------------------
    # C) colunas importantes para preservar
    # -----------------------------
    preserve_cols = [
        "time",
        "Velocidade do veículo (km/h)",
        "RPM do motor (rpm)",
        "RPM do motor (RPM)",  # se vier com caixa diferente
        "Aceleração (g)",
        "Consumo instantâneo de combustível (l/100km) (L/100km)",
        "[ECM] Accelerator pedal: sensor 1 voltage (mV)",
        "[ECM] Accelerator pedal: sensor 2 voltage (mV)",
        "Posição do pedal do acelerador D (%)",
        "Posição do pedal do acelerador E (%)",
    ]

    # manter só as preservadas que realmente existirem
    preserve_cols = [c for c in preserve_cols if c in df.columns]

    # -----------------------------
    # D) remover colunas 100% vazias
    # -----------------------------
    all_nan_cols = df.columns[df.isna().all()].tolist()
    all_nan_cols = [c for c in all_nan_cols if c not in preserve_cols]
    df = df.drop(columns=all_nan_cols, errors="ignore")

    # -----------------------------
    # E) remover colunas muito esparsas
    # -----------------------------
    nan_ratio = df.isna().mean()
    sparse_cols = nan_ratio[nan_ratio > thresh_nan].index.tolist()
    sparse_cols = [c for c in sparse_cols if c not in preserve_cols]
    df = df.drop(columns=sparse_cols, errors="ignore")

    # -----------------------------
    # F) converter tempo
    # -----------------------------
    if "time" in df.columns:
        df["time"] = pd.to_datetime(df["time"], errors="coerce")
        df = df[df["time"].notna()].copy()
        df = df.sort_values("time").drop_duplicates(subset=["time"], keep="last")
        df = df.reset_index(drop=True)

    # -----------------------------
    # G) padronizar nomes principais
    # -----------------------------
    rename_map = {
        "Velocidade do veículo (km/h)": "speed_kmh",
        "RPM do motor (rpm)": "rpm",
        "RPM do motor (RPM)": "rpm",
        "Aceleração (g)": "acc_g",
        "Consumo instantâneo de combustível (l/100km) (L/100km)": "fuel_l_per_100km",
        "[ECM] Accelerator pedal: sensor 1 voltage (mV)": "pedal_sensor1_mV",
        "[ECM] Accelerator pedal: sensor 2 voltage (mV)": "pedal_sensor2_mV",
        "Posição do pedal do acelerador D (%)": "pedal_D_pct",
        "Posição do pedal do acelerador E (%)": "pedal_E_pct",
    }
    rename_map = {k: v for k, v in rename_map.items() if k in df.columns}
    df = df.rename(columns=rename_map)

    # -----------------------------
    # H) converter colunas numéricas relevantes
    # -----------------------------
    cols_numericas = [
        "speed_kmh", "rpm", "acc_g", "fuel_l_per_100km",
        "pedal_sensor1_mV", "pedal_sensor2_mV",
        "pedal_D_pct", "pedal_E_pct"
    ]
    for col in cols_numericas:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # -----------------------------
    # I) criar pedal_real_pct (quando houver D/E)
    #     - mesmo com pouca cobertura, vale salvar
    # -----------------------------
    pedal_cols = [c for c in ["pedal_D_pct", "pedal_E_pct"] if c in df.columns]
    if pedal_cols:
        df["pedal_real_pct"] = df[pedal_cols].mean(axis=1, skipna=True)

    # -----------------------------
    # J) salvar
    # -----------------------------
    df.to_csv(path_out, index=False)

    # -----------------------------
    # K) resumo
    # -----------------------------
    print("=" * 72)
    print("RESUMO - VOLKS COMBUSTÃO")
    print("=" * 72)
    print(f"Shape final: {df.shape}")
    print(f"Arquivo salvo em: {path_out}")

    colunas_resumo = [c for c in [
        "time", "speed_kmh", "rpm", "acc_g", "fuel_l_per_100km",
        "pedal_sensor1_mV", "pedal_sensor2_mV",
        "pedal_D_pct", "pedal_E_pct", "pedal_real_pct"
    ] if c in df.columns]

    print("\nCobertura (% não nulo):")
    for c in colunas_resumo:
        pct = 100 * df[c].notna().mean()
        print(f" - {c}: {pct:.2f}%")

    return df

# =========================================================
# 4) TRATAMENTO DO CSVLOG (arquivo com metadado na 1ª linha)
# =========================================================
def tratar_csvlog(path_in, path_out):
    # skiprows=1 porque a primeira linha é:
    # # StartTime = ...
    df = pd.read_csv(path_in, skiprows=1)

    # -----------------------------
    # A) limpar nomes de colunas
    # -----------------------------
    df.columns = [normalizar_nome_coluna(c) for c in df.columns]

    # -----------------------------
    # B) renomear colunas principais
    # -----------------------------
    rename_map = {
        "Time (sec)": "time_sec",
        "Velocidade do veículo (km/h)": "speed_kmh",
        "RPM do motor (RPM)": "rpm",
        "Posição absoluta do acelerador (%)": "throttle_abs_pct",
        "Posição relativa do acelerador (%)": "throttle_rel_pct",
        "Posição absoluta do acelerador B (%)": "throttle_abs_B_pct",
        "Posição absoluta do acelerador C (%)": "throttle_abs_C_pct",
        "Posição D do pedal do acelerador (%)": "pedal_D_pct",
        "Posição do pedal do acelerador E (%)": "pedal_E_pct",
        "Controle do atuador do acelerador comandado (%)": "throttle_cmd_pct",
        "Economia de combustível instantânea (l/100 km)": "fuel_l_per_100km",
        "Potência do motor (PS)": "engine_power_ps",
        "Torque do motor (N•m)": "engine_torque_nm",
        "Latitude (deg)": "latitude",
        "Longitude (deg)": "longitude",
        "Altitude (m)": "altitude_m",
        "Velocidade do GPS (km/h)": "gps_speed_kmh",
        "Aceleração X (m/s²)": "acc_x_ms2",
        "Acelere Y (m/s²)": "acc_y_ms2",
        "Aceleração Z (m/s²)": "acc_z_ms2",
    }
    rename_map = {k: v for k, v in rename_map.items() if k in df.columns}
    df = df.rename(columns=rename_map)

    # -----------------------------
    # C) converter tudo que puder para numérico
    # -----------------------------
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="ignore")

    # -----------------------------
    # D) garantir ordenação temporal
    # -----------------------------
    if "time_sec" in df.columns:
        df["time_sec"] = pd.to_numeric(df["time_sec"], errors="coerce")
        df = df[df["time_sec"].notna()].copy()
        df = df.sort_values("time_sec").drop_duplicates(subset=["time_sec"], keep="last")
        df = df.reset_index(drop=True)

    # -----------------------------
    # E) criar coluna única de pedal real
    #     - média entre D e E
    # -----------------------------
    pedal_cols = [c for c in ["pedal_D_pct", "pedal_E_pct"] if c in df.columns]
    if pedal_cols:
        df["pedal_real_pct"] = df[pedal_cols].mean(axis=1, skipna=True)

    # -----------------------------
    # F) opcional: coluna de aceleração consolidada
    #     magnitude do vetor (se 3 eixos existirem)
    # -----------------------------
    if all(c in df.columns for c in ["acc_x_ms2", "acc_y_ms2", "acc_z_ms2"]):
        df["acc_magnitude_ms2"] = np.sqrt(
            df["acc_x_ms2"]**2 +
            df["acc_y_ms2"]**2 +
            df["acc_z_ms2"]**2
        )

    # -----------------------------
    # G) salvar
    # -----------------------------
    df.to_csv(path_out, index=False)

    # -----------------------------
    # H) resumo
    # -----------------------------
    print("=" * 72)
    print("RESUMO - CSVLOG COMBUSTÃO")
    print("=" * 72)
    print(f"Shape final: {df.shape}")
    print(f"Arquivo salvo em: {path_out}")

    colunas_resumo = [c for c in [
        "time_sec", "speed_kmh", "rpm",
        "pedal_D_pct", "pedal_E_pct", "pedal_real_pct",
        "throttle_abs_pct", "throttle_rel_pct", "throttle_cmd_pct",
        "fuel_l_per_100km", "engine_power_ps", "engine_torque_nm"
    ] if c in df.columns]

    print("\nCobertura (% não nulo):")
    for c in colunas_resumo:
        pct = 100 * df[c].notna().mean()
        print(f" - {c}: {pct:.2f}%")

    return df

# =========================================================
# 5) RODAR NOS DOIS ARQUIVOS
# =========================================================
df_volks_tratado = tratar_volks(
    path_in=arquivo_volks,
    path_out=saida_volks,
    thresh_nan=0.98   # pode baixar para 0.95 se quiser limpeza mais agressiva
)

print("\n")

df_csvlog_tratado = tratar_csvlog(
    path_in=arquivo_csvlog,
    path_out=saida_csvlog
)

# =========================================================
# 6) PRÉVIA
# =========================================================
print("\n" + "=" * 72)
print("AMOSTRA - VOLKS TRATADO")
print("=" * 72)
print(df_volks_tratado.head())

print("\n" + "=" * 72)
print("AMOSTRA - CSVLOG TRATADO")
print("=" * 72)
print(df_csvlog_tratado.head())


# =========================================================
# 7) BINAR O VOLKS TRATADO PARA ALINHAR SINAIS
# =========================================================
def binar_volks_tratado(df, freq="1s", path_out=None):
    """
    Agrupa o Volks em janelas temporais maiores para fazer
    speed/rpm/fuel/acc caírem na mesma linha com mais frequência.
    """
    data = df.copy()

    if "time" not in data.columns:
        raise ValueError("O DataFrame do Volks não possui a coluna 'time'.")

    data["time"] = pd.to_datetime(data["time"], errors="coerce")
    data = data[data["time"].notna()].copy()
    data = data.sort_values("time").reset_index(drop=True)

    cols_agregar = [
        "speed_kmh",
        "rpm",
        "acc_g",
        "fuel_l_per_100km",
        "pedal_sensor1_mV",
        "pedal_sensor2_mV",
        "pedal_D_pct",
        "pedal_E_pct",
        "pedal_real_pct",
    ]
    cols_agregar = [c for c in cols_agregar if c in data.columns]

    for col in cols_agregar:
        data[col] = pd.to_numeric(data[col], errors="coerce")

    # usa floor para colocar tudo no mesmo "balde" de tempo
    data["time_bin"] = data["time"].dt.floor(freq)

    # média dentro de cada bin
    agg_dict = {col: "mean" for col in cols_agregar}

    df_bin = (
        data.groupby("time_bin", as_index=False)
            .agg(agg_dict)
            .rename(columns={"time_bin": "time"})
    )

    # conta quantos sinais base existem por linha
    base_cols = [c for c in ["speed_kmh", "rpm", "fuel_l_per_100km", "acc_g"] if c in df_bin.columns]
    if base_cols:
        df_bin["n_sinais_base"] = df_bin[base_cols].notna().sum(axis=1)

    if path_out is not None:
        df_bin.to_csv(path_out, index=False)

    print("=" * 72)
    print("RESUMO - VOLKS BINADO")
    print("=" * 72)
    print(f"Frequência usada: {freq}")
    print(f"Shape final: {df_bin.shape}")

    if base_cols:
        print("\nCoexistência dos sinais-base por linha:")
        print(df_bin["n_sinais_base"].value_counts().sort_index())

    print("\nCobertura (% não nulo):")
    for c in cols_agregar:
        pct = 100 * df_bin[c].notna().mean()
        print(f" - {c}: {pct:.2f}%")

    if path_out is not None:
        print(f"\nArquivo salvo em: {path_out}")

    return df_bin

# =========================================================
# 8) GERAR UMA VERSÃO BINADA DO VOLKS
# =========================================================
saida_volks_binado = "volks_combustao_tratado_binado_1s.csv"

df_volks_binado = binar_volks_tratado(
    df=df_volks_tratado,
    freq="500ms",   # teste "500ms", "1s" ou "2s"
    path_out=saida_volks_binado
)

print("\n" + "=" * 72)
print("AMOSTRA - VOLKS BINADO")
print("=" * 72)
print(df_volks_binado.head(10))
```

### Proxy byd e Pegeout



```python
import pandas as pd
import numpy as np

# =========================================================
# 1) CAMINHOS
# =========================================================
arquivo_byd = "byd_dolphin_reamostrado_v2.csv"
arquivo_peugeot = "peugeot_2008_reamostrado_v2.csv"

saida_byd = "byd_dolphin_com_pedal_proxy.csv"
saida_peugeot = "peugeot_2008_com_pedal_proxy.csv"

# =========================================================
# 2) FUNÇÕES AUXILIARES
# =========================================================
def robust_minmax(series, q_low=0.05, q_high=0.95):
    s = pd.to_numeric(series, errors="coerce")
    low = s.quantile(q_low)
    high = s.quantile(q_high)

    if pd.isna(low) or pd.isna(high) or high == low:
        return pd.Series(np.nan, index=s.index)

    out = (s - low) / (high - low)
    return out.clip(0, 1)

def classificar_pedal(x):
    if pd.isna(x):
        return np.nan
    if x < 15:
        return "solto"
    elif x < 40:
        return "leve"
    elif x < 70:
        return "moderado"
    else:
        return "forte"

def criar_pedal_proxy_ev(path_in, path_out, nome="EV"):
    df = pd.read_csv(path_in)

    # ---------------------------------------------
    # A) Garantir colunas esperadas
    # ---------------------------------------------
    cols_necessarias = ["current_A", "power_hp", "acc_g", "speed_kmh"]
    cols_faltando = [c for c in cols_necessarias if c not in df.columns]
    if cols_faltando:
        raise ValueError(f"{nome}: faltam colunas: {cols_faltando}")

    for col in cols_necessarias:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # ---------------------------------------------
    # B) Componentes positivos (tração)
    # ---------------------------------------------
    df["current_pos"] = df["current_A"].clip(lower=0)
    df["power_pos"] = df["power_hp"].clip(lower=0)
    df["acc_pos"] = df["acc_g"].clip(lower=0)

    # delta de velocidade
    df["delta_v"] = df["speed_kmh"].diff()
    df["delta_v_pos"] = df["delta_v"].clip(lower=0)

    # ---------------------------------------------
    # C) Normalização robusta
    # ---------------------------------------------
    df["I_norm"] = robust_minmax(df["current_pos"])
    df["P_norm"] = robust_minmax(df["power_pos"])
    df["A_norm"] = robust_minmax(df["acc_pos"])
    df["DV_norm"] = robust_minmax(df["delta_v_pos"])

    # ---------------------------------------------
    # D) Máscara de validade
    # usa a existente se já houver; senão cria
    # ---------------------------------------------
    if "valid_for_pedal_proxy" in df.columns:
        valid_mask = df["valid_for_pedal_proxy"].fillna(0).astype(int) == 1
    else:
        valid_mask = (
            df["current_A"].notna() &
            df["speed_kmh"].notna() &
            (
                df["power_hp"].notna() |
                df["acc_g"].notna()
            )
        )

    # ---------------------------------------------
    # E) Fórmula do proxy bruto
    # ---------------------------------------------
    df["pedal_proxy_raw"] = (
        0.45 * df["I_norm"].fillna(0) +
        0.30 * df["P_norm"].fillna(0) +
        0.20 * df["A_norm"].fillna(0) +
        0.05 * df["DV_norm"].fillna(0)
    )

    # invalida onde não há base mínima
    df.loc[~valid_mask, "pedal_proxy_raw"] = np.nan

    # ---------------------------------------------
    # F) Suavização temporal
    # ---------------------------------------------
    df["pedal_proxy_smooth"] = (
        df["pedal_proxy_raw"]
        .rolling(window=5, min_periods=1)
        .mean()
    )

    # ---------------------------------------------
    # G) Escala final 0-100
    # ---------------------------------------------
    df["pedal_proxy_pct"] = (100 * df["pedal_proxy_smooth"]).clip(0, 100)

    # classe
    df["pedal_proxy_class"] = df["pedal_proxy_pct"].apply(classificar_pedal)

    # ---------------------------------------------
    # H) Salvar
    # ---------------------------------------------
    df.to_csv(path_out, index=False)

    # ---------------------------------------------
    # I) Resumo
    # ---------------------------------------------
    print("=" * 72)
    print(f"RESUMO - {nome}")
    print("=" * 72)
    print(f"Linhas totais: {len(df)}")
    print(f"pedal_proxy_pct não nulo: {df['pedal_proxy_pct'].notna().sum()} ({100*df['pedal_proxy_pct'].notna().mean():.2f}%)")
    print(f"Mínimo: {df['pedal_proxy_pct'].min():.2f}")
    print(f"Máximo: {df['pedal_proxy_pct'].max():.2f}")
    print(f"Média : {df['pedal_proxy_pct'].mean():.2f}")
    print(f"Arquivo salvo em: {path_out}")

    print("\nDistribuição das classes:")
    print(df["pedal_proxy_class"].value_counts(dropna=False))

    return df

# =========================================================
# 3) RODAR
# =========================================================
df_byd_proxy = criar_pedal_proxy_ev(
    path_in=arquivo_byd,
    path_out=saida_byd,
    nome="BYD Dolphin"
)

print("\n")

df_peugeot_proxy = criar_pedal_proxy_ev(
    path_in=arquivo_peugeot,
    path_out=saida_peugeot,
    nome="Peugeot e-2008"
)

# =========================================================
# 4) PRÉVIA
# =========================================================
print("\n" + "=" * 72)
print("AMOSTRA - BYD COM PROXY")
print("=" * 72)
print(df_byd_proxy[[
    "time", "current_A", "power_hp", "acc_g", "speed_kmh",
    "pedal_proxy_pct", "pedal_proxy_class"
]].head(10))

print("\n" + "=" * 72)
print("AMOSTRA - PEUGEOT COM PROXY")
print("=" * 72)
print(df_peugeot_proxy[[
    "time", "current_A", "power_hp", "acc_g", "speed_kmh",
    "pedal_proxy_pct", "pedal_proxy_class"
]].head(10))
```

```python
import pandas as pd
import numpy as np

# =========================================================
# 1) CAMINHO
# =========================================================
arquivo_volks = "volks_combustao_tratado_binado_1s.csv"
saida_volks = "volks_combustao_com_pedal_proxy.csv"

# =========================================================
# 2) FUNÇÕES AUXILIARES
# =========================================================
def robust_minmax(series, q_low=0.05, q_high=0.95):
    s = pd.to_numeric(series, errors="coerce")
    low = s.quantile(q_low)
    high = s.quantile(q_high)

    if pd.isna(low) or pd.isna(high) or high == low:
        return pd.Series(np.nan, index=s.index)

    out = (s - low) / (high - low)
    return out.clip(0, 1)

def classificar_pedal(x):
    if pd.isna(x):
        return np.nan
    if x < 15:
        return "solto"
    elif x < 40:
        return "leve"
    elif x < 70:
        return "moderado"
    else:
        return "forte"

def criar_pedal_proxy_volks(path_in, path_out, nome="Volks combustão"):
    df = pd.read_csv(path_in)

    # ---------------------------------------------
    # A) Garantir colunas mínimas
    # ---------------------------------------------
    cols_necessarias = ["rpm", "fuel_l_per_100km", "acc_g", "speed_kmh"]
    cols_faltando = [c for c in cols_necessarias if c not in df.columns]
    if cols_faltando:
        raise ValueError(f"{nome}: faltam colunas: {cols_faltando}")

    for col in cols_necessarias:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # ---------------------------------------------
    # B) Componentes positivos
    # ---------------------------------------------
    df["rpm_pos"] = df["rpm"].clip(lower=0)
    df["fuel_pos"] = df["fuel_l_per_100km"].clip(lower=0)
    df["acc_pos"] = df["acc_g"].clip(lower=0)

    df["delta_v"] = df["speed_kmh"].diff()
    df["delta_v_pos"] = df["delta_v"].clip(lower=0)

    # ---------------------------------------------
    # C) Normalização robusta
    # ---------------------------------------------
    df["RPM_norm"] = robust_minmax(df["rpm_pos"])
    df["FUEL_norm"] = robust_minmax(df["fuel_pos"])
    df["A_norm"] = robust_minmax(df["acc_pos"])
    df["DV_norm"] = robust_minmax(df["delta_v_pos"])

    # ---------------------------------------------
    # D) Máscara de validade
    # ---------------------------------------------
    valid_mask = (
        df["rpm"].notna() &
        df["speed_kmh"].notna() &
        (
            df["fuel_l_per_100km"].notna() |
            df["acc_g"].notna()
        )
    )

    # ---------------------------------------------
    # E) Proxy bruto
    # ---------------------------------------------
    df["pedal_proxy_raw"] = (
        0.35 * df["RPM_norm"].fillna(0) +
        0.35 * df["FUEL_norm"].fillna(0) +
        0.20 * df["A_norm"].fillna(0) +
        0.10 * df["DV_norm"].fillna(0)
    )

    df.loc[~valid_mask, "pedal_proxy_raw"] = np.nan

    # ---------------------------------------------
    # F) Suavização
    # ---------------------------------------------
    df["pedal_proxy_smooth"] = (
        df["pedal_proxy_raw"]
        .rolling(window=5, min_periods=1)
        .mean()
    )

    # ---------------------------------------------
    # G) Escala final
    # ---------------------------------------------
    df["pedal_proxy_pct"] = (100 * df["pedal_proxy_smooth"]).clip(0, 100)
    df["pedal_proxy_class"] = df["pedal_proxy_pct"].apply(classificar_pedal)

    # ---------------------------------------------
    # H) Se houver pedal real, manter para comparação
    # ---------------------------------------------
    if "pedal_real_pct" in df.columns:
        df["pedal_real_pct"] = pd.to_numeric(df["pedal_real_pct"], errors="coerce")

    # ---------------------------------------------
    # I) Salvar
    # ---------------------------------------------
    df.to_csv(path_out, index=False)

    # ---------------------------------------------
    # J) Resumo
    # ---------------------------------------------
    print("=" * 72)
    print(f"RESUMO - {nome}")
    print("=" * 72)
    print(f"Linhas totais: {len(df)}")
    print(f"pedal_proxy_pct não nulo: {df['pedal_proxy_pct'].notna().sum()} ({100*df['pedal_proxy_pct'].notna().mean():.2f}%)")
    print(f"Mínimo: {df['pedal_proxy_pct'].min():.2f}")
    print(f"Máximo: {df['pedal_proxy_pct'].max():.2f}")
    print(f"Média : {df['pedal_proxy_pct'].mean():.2f}")

    if "pedal_real_pct" in df.columns:
        print(f"pedal_real_pct não nulo: {df['pedal_real_pct'].notna().sum()} ({100*df['pedal_real_pct'].notna().mean():.2f}%)")

    print(f"Arquivo salvo em: {path_out}")

    print("\nDistribuição das classes:")
    print(df["pedal_proxy_class"].value_counts(dropna=False))

    return df

# =========================================================
# 3) RODAR
# =========================================================
df_volks_proxy = criar_pedal_proxy_volks(
    path_in=arquivo_volks,
    path_out=saida_volks,
    nome="Volks combustão"
)

# =========================================================
# 4) PRÉVIA
# =========================================================
cols_preview = [
    c for c in [
        "time", "speed_kmh", "rpm", "fuel_l_per_100km", "acc_g",
        "pedal_real_pct", "pedal_proxy_pct", "pedal_proxy_class"
    ] if c in df_volks_proxy.columns
]

print("\n" + "=" * 72)
print("AMOSTRA - VOLKS COM PROXY")
print("=" * 72)
print(df_volks_proxy[cols_preview].head(10))
```

```python
import pandas as pd
import numpy as np

# =========================================================
# 1) ENTRADAS
# =========================================================
arquivo_byd = "byd_dolphin_com_pedal_proxy.csv"
arquivo_peugeot = "peugeot_2008_com_pedal_proxy.csv"
arquivo_volks = "volks_combustao_com_pedal_proxy.csv"
arquivo_csvlog = "csvlog_combustao_tratado.csv"

# =========================================================
# 2) SAÍDAS
# =========================================================
saida_byd = "byd_ev_ml_ready.csv"
saida_peugeot = "peugeot_ev_ml_ready.csv"
saida_volks = "volks_ice_ml_ready.csv"
saida_csvlog = "csvlog_ice_ml_ready.csv"

# =========================================================
# 3) FUNÇÕES AUXILIARES
# =========================================================
def to_numeric_if_exists(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def add_missing_columns(df, cols):
    for c in cols:
        if c not in df.columns:
            df[c] = np.nan
    return df

def remove_constant_columns(df, ignore_cols=None):
    ignore_cols = ignore_cols or []
    remove_cols = []
    for c in df.columns:
        if c in ignore_cols:
            continue
        if df[c].nunique(dropna=True) <= 1:
            remove_cols.append(c)
    return df.drop(columns=remove_cols, errors="ignore"), remove_cols

def ensure_time_columns(df):
    """
    Cria duas colunas padronizadas de tempo:
    - time_ref: datetime quando houver coluna 'time'
    - time_seconds: segundos relativos ao início (sempre que possível)
    """
    if "time" in df.columns:
        t = pd.to_datetime(df["time"], errors="coerce")
        df["time_ref"] = t
        if t.notna().any():
            t0 = t.dropna().iloc[0]
            df["time_seconds"] = (t - t0).dt.total_seconds()
        else:
            df["time_seconds"] = np.nan

    elif "time_sec" in df.columns:
        ts = pd.to_numeric(df["time_sec"], errors="coerce")
        df["time_seconds"] = ts
        df["time_ref"] = pd.NaT

    else:
        df["time_seconds"] = np.nan
        df["time_ref"] = pd.NaT

    return df

# =========================================================
# 4) ESQUEMAS PADRÃO
# =========================================================
EV_FEATURES = [
    "time_ref",
    "time_seconds",
    "speed_kmh",
    "acc_g",
    "current_A",
    "voltage_V",
    "power_hp",
    "gps_speed_kmh",
    "gps_avg_speed_kmh",
    "gps_altitude_m",
    "latitude",
    "longitude",
    "delta_speed_kmh",
    "is_original_sample",
]

ICE_FEATURES = [
    "time_ref",
    "time_seconds",
    "speed_kmh",
    "rpm",
    "acc_g",
    "fuel_l_per_100km",
    "pedal_sensor1_mV",
    "pedal_sensor2_mV",
    "throttle_abs_pct",
    "throttle_rel_pct",
    "throttle_cmd_pct",
    "engine_power_ps",
    "engine_torque_nm",
    "delta_speed_kmh",
    "n_sinais_base",
]

META_COLS = [
    "target_pedal_pct",
    "vehicle_group",
    "vehicle_name",
]

# =========================================================
# 5) PREPARAR ELÉTRICOS
# =========================================================
def preparar_ev(path_in, path_out, vehicle_name):
    df = pd.read_csv(path_in)
    df = ensure_time_columns(df)

    numeric_cols = [
        "time_seconds",
        "speed_kmh", "acc_g", "current_A", "voltage_V", "power_hp",
        "gps_speed_kmh", "gps_avg_speed_kmh", "gps_altitude_m",
        "latitude", "longitude", "pedal_proxy_pct", "is_original_sample",
        "valid_for_pedal_proxy"
    ]
    df = to_numeric_if_exists(df, numeric_cols)

    if "pedal_proxy_pct" not in df.columns:
        raise ValueError(f"{vehicle_name}: 'pedal_proxy_pct' não encontrada.")

    df["target_pedal_pct"] = df["pedal_proxy_pct"]

    if "speed_kmh" in df.columns:
        df["delta_speed_kmh"] = pd.to_numeric(df["speed_kmh"], errors="coerce").diff()

    df = df[df["target_pedal_pct"].notna()].copy()

    if "valid_for_pedal_proxy" in df.columns:
        df = df[df["valid_for_pedal_proxy"].fillna(0).astype(int) == 1].copy()

    keep_cols = [c for c in EV_FEATURES if c in df.columns] + ["target_pedal_pct"]
    df = df[keep_cols].copy()

    df = add_missing_columns(df, EV_FEATURES + META_COLS)
    df["vehicle_group"] = "EV"
    df["vehicle_name"] = vehicle_name

    final_cols = EV_FEATURES + META_COLS
    df = df[final_cols].copy()

    # Não remover colunas de tempo, mesmo que tenham baixa variância
    df, removed = remove_constant_columns(
        df,
        ignore_cols=["time_ref", "time_seconds", "target_pedal_pct", "vehicle_group", "vehicle_name"]
    )

    df = add_missing_columns(df, final_cols)
    df = df[final_cols].copy()

    df.to_csv(path_out, index=False)

    print("=" * 72)
    print(f"EV FINALIZADO - {vehicle_name}")
    print("=" * 72)
    print(f"Shape final: {df.shape}")
    print(f"Arquivo salvo em: {path_out}")
    print(f"Colunas constantes removidas: {removed if removed else 'Nenhuma'}")
    print("\nCobertura (% não nulo):")
    for c in final_cols:
        print(f" - {c}: {100 * df[c].notna().mean():.2f}%")

    return df

# =========================================================
# 6) PREPARAR COMBUSTÃO
# =========================================================
def preparar_ice(path_in, path_out, vehicle_name, usar_proxy=False):
    df = pd.read_csv(path_in)
    df = ensure_time_columns(df)

    numeric_cols = [
        "time_seconds",
        "speed_kmh", "rpm", "acc_g", "fuel_l_per_100km",
        "pedal_sensor1_mV", "pedal_sensor2_mV",
        "throttle_abs_pct", "throttle_rel_pct", "throttle_cmd_pct",
        "engine_power_ps", "engine_torque_nm",
        "pedal_proxy_pct", "pedal_real_pct", "n_sinais_base", "acc_magnitude_ms2"
    ]
    df = to_numeric_if_exists(df, numeric_cols)

    if usar_proxy:
        if "pedal_proxy_pct" not in df.columns:
            raise ValueError(f"{vehicle_name}: 'pedal_proxy_pct' não encontrada.")
        df["target_pedal_pct"] = df["pedal_proxy_pct"]
    else:
        if "pedal_real_pct" not in df.columns:
            raise ValueError(f"{vehicle_name}: 'pedal_real_pct' não encontrada.")
        df["target_pedal_pct"] = df["pedal_real_pct"]

    if "speed_kmh" in df.columns:
        df["delta_speed_kmh"] = pd.to_numeric(df["speed_kmh"], errors="coerce").diff()

    df = df[df["target_pedal_pct"].notna()].copy()

    if "n_sinais_base" in df.columns:
        df = df[df["n_sinais_base"].fillna(0) >= 3].copy()

    if ("acc_g" not in df.columns or df["acc_g"].isna().all()) and "acc_magnitude_ms2" in df.columns:
        df["acc_g"] = pd.to_numeric(df["acc_magnitude_ms2"], errors="coerce") / 9.80665

    keep_cols = [c for c in ICE_FEATURES if c in df.columns] + ["target_pedal_pct"]
    df = df[keep_cols].copy()

    df = add_missing_columns(df, ICE_FEATURES + META_COLS)
    df["vehicle_group"] = "ICE"
    df["vehicle_name"] = vehicle_name

    final_cols = ICE_FEATURES + META_COLS
    df = df[final_cols].copy()

    df, removed = remove_constant_columns(
        df,
        ignore_cols=["time_ref", "time_seconds", "target_pedal_pct", "vehicle_group", "vehicle_name"]
    )

    df = add_missing_columns(df, final_cols)
    df = df[final_cols].copy()

    df.to_csv(path_out, index=False)

    print("=" * 72)
    print(f"ICE FINALIZADO - {vehicle_name}")
    print("=" * 72)
    print(f"Shape final: {df.shape}")
    print(f"Arquivo salvo em: {path_out}")
    print(f"Colunas constantes removidas: {removed if removed else 'Nenhuma'}")
    print("\nCobertura (% não nulo):")
    for c in final_cols:
        print(f" - {c}: {100 * df[c].notna().mean():.2f}%")

    return df

# =========================================================
# 7) EXECUTAR
# =========================================================
df_byd_ev = preparar_ev(
    path_in=arquivo_byd,
    path_out=saida_byd,
    vehicle_name="BYD_Dolphin"
)

print("\n")

df_peugeot_ev = preparar_ev(
    path_in=arquivo_peugeot,
    path_out=saida_peugeot,
    vehicle_name="Peugeot_e2008"
)

print("\n")

df_volks_ice = preparar_ice(
    path_in=arquivo_volks,
    path_out=saida_volks,
    vehicle_name="Volks_Combustao",
    usar_proxy=True
)

print("\n")

df_csvlog_ice = preparar_ice(
    path_in=arquivo_csvlog,
    path_out=saida_csvlog,
    vehicle_name="CSVLog_Combustao",
    usar_proxy=False
)

# =========================================================
# 8) PRÉVIAS
# =========================================================
print("\n" + "=" * 72)
print("AMOSTRA - BYD EV ML READY")
print("=" * 72)
print(df_byd_ev.head())

print("\n" + "=" * 72)
print("AMOSTRA - PEUGEOT EV ML READY")
print("=" * 72)
print(df_peugeot_ev.head())

print("\n" + "=" * 72)
print("AMOSTRA - VOLKS ICE ML READY")
print("=" * 72)
print(df_volks_ice.head())

print("\n" + "=" * 72)
print("AMOSTRA - CSVLOG ICE ML READY")
print("=" * 72)
print(df_csvlog_ice.head())
```
